```python
#From PMS classes and objects 1
class point:
    pass

blank=point()
blank.x=3.0    #x and y are the attributes of blank object
blank.y=4.0
print(blank.y) 
x=blank.x      # Here x is a variable
print(x)       #printing variable x

print("("+str(blank.x)+","+str(blank.y)+")")
print("(",blank.x,",",blank.y,")")
distancesquared = blank.x*blank.x + blank.y*blank.y
print(distancesquared)
print (blank)     #blank object declared in main at address xxHexxxxx

#instances as arguments
def printPoint(p):
    print("("+str(p.x)+","+str(p.y)+")")
printPoint(blank)

#sameness
p1=point()  #Another object of class point
p1.x=3
p1.y=4
p2=point()
p2.x=3
p2.y=4
p2.z=6
print(p1==p2)  #Result false
print(p2.z)

#Shallow Equality = comparing only refrences of object not the contents of the object
p1=p2 #Assigning p1 and p2 as aliases of each other
print(p2.z)   #p2.z, z is an extra attribute of object p2, still p1==p2 comes true as refrences are checked only, not the contents
print(p1==p2)  #Result True
```

```python
#Deep equality = when compare contents of the object
def samePoint(point1,point2): 
    return (point1.x==point2.x) and (point1.y==point2.y)  #using and operator return (true and true)= true

p3=point()
p3.x,p3.y=8,10
p4=point()
p4.x,p4.y=8,10

p5=point()
p5.x,p5.y=8,12
print(p4==p5)
p4=p5 #p4 as alias of p5
print(p4.y,p5.y)
print(p4==p5)
print(samePoint(p4,p5))
print(samePoint(p3,p4))  #here they have both deep and shallow copy

class Rectangle:
    pass
box=Rectangle()
box.width=100.0
box.height=200.0
box.corner=point()  #box.corner is a point type object
box.corner.x=0.0
box.corner.y=0.0

def findCenter(boxInfo):
    p=point()
    p.x=box.corner.x+box.width/2.0
    p.y=box.corner.y-box.height/2.0
    return p
center = findCenter(box)    #center gets assigned to point type object "p"
printPoint(center)          #printPoint takes point type object "p" as argument    
                            #output= (50.0,-100.0), -100 bcoz it goes down from 0,0
#objects are mutable
#later complete it

```

```python
#Private and public data members in a class
class demo:
    count=0
    def __init__(self,b=0,c=0):
        self.b=b       #public
        self.__c=c     #private
        
    demo.count+=1
    
    def disp(self):
        print(self.b,self.__c,demo.count)
    def __disp1(self):
        print(self.b,self.__c,demo.count)
obj1=demo(10,20)
obj1.disp()
print(obj1.b, obj1._demo__c)
obj1._demo__disp1()
```
