```python
def random_line(file,randline):
    f=open(file)
    lines=f.readlines()
    print(lines)
    f.close()
    if randline>len(lines):
        return "line doesn't exists"
    else:
        return lines[randline-1]

print(random_line("C:\\Users\\hp\\Desktop\\kbtest.txt",3))
```

```python
def count(file):
    f=open(file)
    sc=0
    for i in f.read():
        if i==" " or i=="\n":
            sc+=1
        elif i=="":
            break
    f.close()
    print(sc+1)
count("K20TS.txt")


```

```python
def count(file):
    f=open(file)
    c=f.read()
    print(c)
    f.close()
    if c=="":
        return "file is empty"
    else:
        l=c.split()
        print(l)
        return len(l)
print(count("K20TS.txt"))
```
