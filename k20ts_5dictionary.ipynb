```python
#Ceaser cipher using dictionary
d={}
d1={}
for i in range(0,26):
    d[i]=chr(97+i)
    d1[chr(97+i)]=i
#print(d,d1)
```

```python
def encryption(s="",k=1):
    s1=""
    for i in s:
        if i.isalpha():
            if(d1[i]+k<=25):
                s1+=d[d1[i]+k]
            else:
                s1+=d[(d1[i]+k)%26]
        else:
            s1+=i
    return s1

print(encryption("hello",4))
print(encryption("xyz",3))

            
        
```
