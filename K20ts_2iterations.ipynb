```python
def kangaroo(x1, v1, x2, v2):
    # Write your code here
    for i in range(0,10001):
        if(x1+i*v1==x2+i*v2): 
            return "YES"
    return "NO"

result = kangaroo(0,3,4,2)
print(result)
```

```python
#Reverse of a number
num=int(input("Enter a number"))
r,n=0,0
while(num>0):
    r=num%10
    n=n*10+r
    num=num//10
print(n)
```

```python
num = int(input('Enter a number '))
for i in range(0, num):
    print(' ' * (num - i), '*', end='')
    if i == num//2:
        print((num + 2) * '*')
        continue
    print(' ' * 2 * i, '*')
```
