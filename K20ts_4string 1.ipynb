```python
def find(s,ch):
    for i in range(len(s)):
        if s[i]==ch:
            return i
    return -1
s1="Welcome to python"
print(find(s1,'e'))
print(find(s1,'f'))

#WAP to make a find function for string where we can 
#search for a substring
```

```python
def findsub(s,sub):
    for i in range(len(s)):
        if(s[i]==sub[0]):
            for j in range(len(sub)):
                if s[i+j]!=sub[j]:
                    break
            else:
                return i
    else:
        return -1

print(findsub("Welcome to python","come"))
print(findsub("come  run  cone", "run"))
print(findsub("heelo","EE"))
```

```python
def duplicates(l):
    i=0
    while(i!=(len(l)-1)):
            if(l[i]==l[i+1]):
                del l[i+1]
            else:
                i+=1
l=[1,1,2,3,4,5,5,6,6,6,1,1,2,2]
duplicates(l)
print(l)
```

```python
'''write down a program to count total number of lowercase and 
total number of uppercase letter in a string'''
def count(s):
    lc,uc=0,0
    for i in s:
        if i in string.ascii_uppercase:
            uc+=1
        if i in string.ascii_lowercase:
            lc+=1
    return ("lowercase "+str(lc)+" Upper case "+str(uc))
s=input("Enter a string")
print(count(s))
```
