```python
# if elif else
a=int(input("Enter a number "))
b=int(input("Enter a number "))
c=int(input("Enter a number "))
if(a>=b and a>c):
    if(a>b):
        print(a, "is greater")
    else:
        print("a and b both are same and greater",a)
elif(b>a and b>=c):
    if(b>c):
        print(b, "is greater")
    else:
        print("c and b both are same and greater",b)
elif(c>=a and c>b):
    if(c>a):
        print(c, "is greater")
    else:
        print("a and c both are same and greater",a)
else:
    print("All are equal")
```
