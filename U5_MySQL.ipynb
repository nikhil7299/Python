```python
db=mysql.connector.connect(
host="localhost",
user='root',
password='root',
database="k20ts")
```

```python
cur=db.cursor()
cur.execute("CREATE TABLE student(rollno INT PRIMARY KEY, name VARCHAR(255), address VARCHAR(255))")
```

```python
cur.execute("SHOW TABLES")
for i in cur:
    print(i)
```

```python
cur.execute("INSERT INTO student VALUES(1,'hello','hi')")
```

```python
cur.execute("INSERT INTO student VALUES(%s,%s,%s)",(2,"a","b"))
```

```python
cur.execute("SELECT * FROM student")
data=cur.fetchall()
for i in data:
    print(i)
```
