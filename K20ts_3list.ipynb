```python
#Q. to find difference btw 2 string lists
def diff(l1,l2):
    l=[]
    for i in l1:
        if i not in l2:
            l.append(i)
    return(l)
l1=['green','yellow','blue','orange','white']
l2=['purple','black','yellow','orange']
print(diff(l1,l2))
print(diff(l2,l1))
```
