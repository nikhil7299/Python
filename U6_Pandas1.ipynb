```python
from tkinter import *
from tkinter import messagebox
import tkmacosx

def insert():
    if(var1.get()=="" or var2.get()=="" or var3.get()=="" or v1.get()==0):
        messagebox.showerror("Error","All fields are required")
    else:
        if(var2.get().isdigit() and len(var2.get())==10):
            import re
            res=re.search(".+@.+['.'].+",var3.get())
            if(res):
                df=pd.read_csv("regis.csv")
                cn=list(df.columns)
                df=df.append(pd.DataFrame({cn[0]:[var1.get()],
                                          cn[1]:[var2.get()],
                                          cn[2]:[var3.get()],
                                          cn[3]:[v1.get()]}))
                df.to_csv("regis.csv",index=None)
                
            else:
                messagebox.showerror("Email","Enter a valid email")
        else:
            messagebox.showerror("Mobile","Enter Valid Mobile Number")


top=Tk(className="Registration")
top.geometry("400x400")
Label(top,text="Enter Name").place(x=50,y=50)
var1=StringVar()
e1=Entry(top,textvariable=var1)
e1.place(x=150,y=50)
Label(top,text="Enter Mobile").place(x=50,y=100)
var2=StringVar()
e2=Entry(top,textvariable=var2)
e2.place(x=150,y=100)
Label(top,text="Enter Email").place(x=50,y=150)
var3=StringVar()
e3=Entry(top,textvariable=var3)
e3.place(x=150,y=150)
Label(top,text="Gender").place(x=50,y=200)
v1=IntVar()
r1=Radiobutton(top,text="Male",variable=v1,value=1)
r1.place(x=100,y=200)
r2=Radiobutton(top,text="Female",variable=v1,value=2)
r2.place(x=180,y=200)
r3=Radiobutton(top,text="Others",variable=v1,value=3)
r3.place(x=260,y=200)
Button(top,text="Submit",command=insert).place(x=150,y=250)
top.mainloop()
```

```python
# Sliding Windows technique
def birthday(s, d, m):
    # Write your code here
    # int sumDay=0
    n=len(s)
    cnt=0
    for i in range(0,n-m+1):
        sumDay=0
        # print(sumDay)
        for j in range(i,i+m):
            sumDay+= s[j]
            # print(sumDay)
        if sumDay==d:
            cnt+=1
    return cnt 

s=[2,2,1,3,2]
d=4
m=2
result=birthday(s, d, m)
print(result)
```
