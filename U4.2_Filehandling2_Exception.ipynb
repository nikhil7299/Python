```python
def random_line(f,line):
    file=open(f)
    l=file.readlines()
    file.close()
    print(l)
    if line>len(l):
        return "line doesn't exist"
    else:
        return l[line-1]
print(random_line("C:/Users/hp/Desktop/kb test.txt",7))
```

```python
#WAP to copy contents of one file into another file line by line
def copy(oldfile, newfile):
    f1=open(oldfile,'r')
    f2=open(newfile,'w')
    while True:
        r=f1.readline()
        if(r==''):
            break
        f2.write(r)
    f1.close()
    f2.close()
print(copy("k20vl.txt",'abc.txt'))
```

```python
def count(file):
    f=open(file)
    l=f.readlines()
    print(l)
    if len(l)==0:
        return "Empty file"
    else:
        w=0
        for i in l:
            w+=len(i.split())
        return w
print(count("k20vl.txt"))
```
