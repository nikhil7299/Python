```python
from tkinter import *
from tkinter import messagebox
import mysql.connector

mydb = mysql.connector.connect(
  host="localhost",
  user="gurneet",
  password="admin",
  database="k20ts"
)

cur=mydb.cursor(buffered=True)

def update():
    if (v2.get()=='' or v3.get()==''):
        messagebox.showerror("ERROR","ALL FIELDS NEED TO BE FILLED")
    else:
        qr="UPDATE studentdata SET name=%s , msg=%s WHERE rno=%s"
        cur.execute(qr,(v2.get(),v3.get(),int(v1.get())))
        mydb.commit()
        messagebox.showinfo("SUCCESS","DATA UPDATED")


def check():
    if (v1.get().isdigit()):
        Label(top,text="DATA FOR THE ROLL NUMBER IS :: ").grid(row=2,column=0)
        qr1="SELECT name FROM studentdata where rno=%s"
        val=v1.get()
        cur.execute(qr1,(val,))
        data1=cur.fetchone()
        qr2="SELECT msg FROM studentdata where rno=%s"
        cur.execute(qr2,(val,))
        data2=cur.fetchone()
        Label(top,text=data1).grid(row=3,column=0)
        Label(top,text=data2).grid(row=4,column=0)
        
        Label(top,text="Enter the new data to be updated :: ").grid(row=7,column=0)
        Label(top,text="Enter Name").grid(row=8,column=0)
        e2=Entry(top,textvariable=v2).grid(row=8,column=1)
        Label(top,text="Enter Message").grid(row=9,column=0)
        e3=Entry(top,textvariable=v3).grid(row=9,column=1)
        Button(top,text="Update Data",command=update).grid(row=10,column=0)
    else:
        messagebox.showerror("ERROR","ROLL NUMBER INCORRECT")

top=Tk()
v2=StringVar()
v3=StringVar()
Label(top,text="Enter roll number").grid(row=0,column=0)
v1=StringVar()
e1=Entry(top,textvariable=v1).grid(row=0,column=1)
Button(top,text="Check Data",command=check).grid(row=1,column=0)
top.mainloop()
```
