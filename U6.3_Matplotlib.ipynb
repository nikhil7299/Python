```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,20)
y=2*x
y1=x**3
plt.plot(x,y,'pk')
plt.plot(x,y1,'-.c')
plt.xlabel("x axis")
plt.ylabel("Y axis")
plt.title("Striaght line")
plt.show()
```

```python
plt.subplot(221)
plt.plot(x,y)
plt.subplot(224)
plt.plot(x,y1)
plt.subplot(223)
y2=np.sin(x)
plt.plot(x,y2)
plt.show()
```
